```python
# Programme description:
# v3_RLH_NB_BMS_seperate_files -
# 1. Daily calculations
# 2. Monthly calculations
# 3. Meter details and daily calculation dataframe merged
# 4. Meter details, daily and monthly dataframe merged
# Written by BOR 06/02/2015

import pandas as pd
import glob
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
pd.options.display.mpl_style = 'default'
from numpy import genfromtxt
from datetime import datetime
import datetime as DT
import os.path
import os

####Daily calculations

# Raw data import - must change location of raw files
ip_dir =r'Z:/03.0 - Environmental/FS Energy Team (Barts Health)/EEC/Python/RLH NB BMS/Raw data June 2014 test'
allFiles = glob.glob(ip_dir + "/*.csv")

# Data export - must change location of export file
out_dir = 'Z:/03.0 - Environmental/FS Energy Team (Barts Health)/EEC/Python/RLH NB BMS/Output June 2014 test'
outfile = out_dir + '/v3_test_5_last_first.csv'

# Meter details, locations, properties
meter_details = 'Z:/03.0 - Environmental/FS Energy Team (Barts Health)/EEC/Python/RLH NB BMS/Meter details/MCC_locations_items.csv'

big_frame = pd.DataFrame()
dfs = []

for file in allFiles:
    rlhnb_bms_merged = pd.read_csv(file, parse_dates=['Date/Time'], dayfirst=True, infer_datetime_format=True, index_col=['Date/Time'], thousands=',') #===> Our raw file
    #print rlhnb_bms_merged.index
    router = str(file[-13:-9])
    
    ##rlhnb_bms_merged = rlhnb_bms_merged['2014-11-05 00:00:00':'2015-01-25 23:30:00']
    rlhnb_bms_merged = rlhnb_bms_merged.convert_objects(convert_numeric=True)
    rlhnb_bms_merged = rlhnb_bms_merged.sort(ascending = True)
    #print rlhnb_bms_merged.dtypes 
    
    # Group by date
    ##rlhnb_bms_merged = rlhnb_bms_merged.groupby(lambda x: x.date()) 
    #print rlhnb_bms_merged.first().head(5)
    #print rlhnb_bms_merged.last().head(5)
    ##first = rlhnb_bms_merged.first()
    ##last = rlhnb_bms_merged.last()
    
    daily_kwh = lambda x: x.iloc[-1] - x.iloc[0]
    #print daily_kwh
    
    transformed = rlhnb_bms_merged.groupby(lambda x: x.date()).apply(daily_kwh)
    #print transformed.dtypes
    pivot = transformed.T  
    #print pivot.dtypes
    pivot['MCC'] = router
    pivot.append(pivot['MCC'])
    pivot.index.name = 'Meter'
    pivot = pivot.reset_index(level=None)
    
    #print pivot_monthly   
    dfs.append(pivot)
            
#===>Concatenate all data into one big DataFrame/daily amounts
big_frame = pd.concat(dfs)

print big_frame.head()

#big_frame.to_csv(outfile, float_format="%.4f")
```

```python
# Data export - must change location of export file
out_dir = 'Z:/03.0 - Environmental/FS Energy Team (Barts Health)/EEC/Python/RLH NB BMS/Output June 2014 test'
outfile = out_dir + '/v3_test_6_last_first.csv'

big_frame_monthly = pd.DataFrame()
dfs_monthly = []

for file in allFiles:
    rlhnb_bms_merged = pd.read_csv(file, parse_dates=['Date/Time'], dayfirst=True, infer_datetime_format=True, index_col=['Date/Time'], thousands=',') #===> Our raw file
    router = str(file[-13:-9])
    ##rlhnb_bms_merged = rlhnb_bms_merged['2014-11-05 00:00:00':'2015-01-25 23:30:00']
    rlhnb_bms_merged = rlhnb_bms_merged.convert_objects(convert_numeric=True)
    rlhnb_bms_merged = rlhnb_bms_merged.sort(ascending = True)
    
    daily_kwh = lambda x: x.iloc[-1] - x.iloc[0]
    
    transformed = rlhnb_bms_merged.groupby(lambda x: x.date()).apply(daily_kwh)
    
    monthly_total = transformed.sum()
    
    pivot_monthly = monthly_total.T  

    pivot_monthly.index.name = 'Meter' 
    pivot_monthly = pivot_monthly.reset_index(level=None)
    pivot_monthly['MCC'] = router
    dfs_monthly.append(pivot_monthly)
    
    
#===>Concatenate all data into one big DataFrame/daily amounts
big_frame_monthly = pd.concat(dfs_monthly)

print big_frame_monthly.head()

big_frame_monthly.to_csv(outfile, float_format="%.4f")
```
